```python
df = pd.read_csv(r"/kaggle/input/top-1000-wealthiest-people-in-the-world/Top_1000_wealthiest_people.csv")
```

```python
sns.histplot(data=df,x="Net Worth (in billions)",bins=30,kde=True)
plt.show()
```

```python
df["Country"].value_counts()
```

```python
ax = sns.countplot(data=df,x="Country",)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
top_country = df["Country"].value_counts().head(10)
sns.barplot(x=top_country.values,y=top_country.index)
plt.title("Tpo Country")
plt.show()
```

```python
df["Industry"].value_counts()
```

```python
ax = sns.countplot(data=df,x="Industry")
plt.xticks(rotation=90)
ax.bar_label(ax.containers[0])
plt.title("top Industrys")
plt.show()
```

```python
df["Company"].value_counts()
```

```python
top_Company = df["Company"].value_counts().head(15)
sns.barplot(x=top_Company.values,y=top_Company.index)
plt.title("Tpo Companys")
plt.show()
```

```python
df["Name"].value_counts()
```

```python
top_Name = df["Name"].value_counts().head(10)
sns.barplot(x=top_Name.values,y=top_Name.index)
plt.title("Count Names")
plt.show()
```

```python
import plotly.express as px

fig = px.sunburst(df, path=['Industry', 'Country'], values='Net Worth (in billions)', 
                  color='Net Worth (in billions)', 
                  color_continuous_scale='RdYlBu', 
                  title='Sunburst Plot of Industry and Country Distribution')
fig.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(x='Industry', y='Country', size='Net Worth (in billions)', hue='Net Worth (in billions)', 
                data=df, palette='cool', sizes=(20, 2000), legend=False)
plt.xticks(rotation=90)
plt.title('Bubble Plot of Net Worth by Industry and Country')
plt.show()
```

```python
industry_count = df['Industry'].value_counts()
plt.figure(figsize=(12, 8))
plt.pie(industry_count, labels=industry_count.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Donut Chart of Industry Representation')
plt.axis('equal')
plt.show()
```
